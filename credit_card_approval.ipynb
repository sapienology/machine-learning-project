```python
# check for nan values and substitute them
print('Uniques values for object columns:')
for nan in df_replace.columns:
  if df_replace[nan].dtype == 'object':
    print(f"column {nan} : {df_replace[nan].unique()} ")
```

replace nan values with the most frequent value repeated in a single column

```python
if df_replace[nan].dtype == 'object':
  frequent = df_replace[nan].value_counts().index[0]
```

```python
# impute missing values using direct assignment
for nan in df_replace.columns:
  if df_replace[nan].dtype == 'object':
    most_frequent_value = df_replace[nan].value_counts().index[0]
    df_replace[nan] = df_replace[nan].fillna(most_frequent_value)
  else:
    mean_value = df_replace[nan].mean()
    df_replace[nan] = df_replace[nan].fillna(mean_value)

```

```python
df_replace = pd.get_dummies(df_replace , drop_first = True)

display(df_replace.head())
```

```python
from sklearn.model_selection import train_test_split

df_replaced = df_replace.values

X,y = df_replaced[: ,:-1] , df_replaced[:,-1]

X_train ,X_test ,y_train , y_test = train_test_split(X,y , test_size = 0.33 , random_state = 42)

y = y.astype(bool)

print(f" training set shape :  {X_train.shape}")
print(f" testing set shape :  {X_test.shape}")
```
